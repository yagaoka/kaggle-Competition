```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(10, 4))
```

```python
df.iloc[1, 0] = np.nan
df.iloc[2:3, 2] = np.nan
df.iloc[5:, 3] = np.nan
```

```python
df
```

```python
df.dropna()
```

```python
df[[0, 1]].dropna()
```

```python
df.fillna(0)
```

```python
df.fillna(method="ffill")
```

```python
df.mean()
```

```python
df.fillna(df.mean())
```

```python
df2 = pd.DataFrame(np.random.rand(15, 6))

df2.iloc[2, 0] = np.nan
df2.iloc[5:9, 2] = np.nan
df2.iloc[7:10, 3] = np.nan
df2.iloc[10, 5] = np.nan

df2
```

```python
df2.dropna()
```

```python
df2.fillna(0)
```

```python
df2.fillna(df2.mean())
```

```python
df2.fillna(df2.median())
```
